## Which number is positive or negative 

```python
num1 = int(input('Enter the first number'))
if num1  > 0:
    print('Number is positive')
elif num1 < 0:
    print('Number is Negative')
else:
    print('Number is zero')
```

## Which number is even or odd

```python
num = int(input('Enter the number'))
if num % 2 == 0:
    print("Number is even")
else:
    print('Number is odd')
```

## Who students passed the examination .

```python
student_marks = int(input("Enter student marks: "))
if student_marks >= 40:
    print("Student has passed the examination.")
else:
    print("Student has failed the examination")
```

## Who is eligible for vote the election 

```python
citizen_age = int(input("Enter the age of the citizen: ")) 
if citizen_age >= 18:
    print('Citizen is eligible to vote in the election.')
else:
    print('Citizen is not eligible to vote in the election.')
```

## Leap Year

```python
year = int(input("Enter a year: "))
if year%400==0:
    print(f"{year} is a leap year") # f string 
elif year%100==0:
    print(f"{year} is not a leap year")
elif year%4==0:
    print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year")
```
